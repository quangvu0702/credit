```python
#export

import random
import pandas as pd
import numpy as np


def shuffle_dataset(n, s=30, ratial=80):
    start_valid = int((n / 100) * ratial)
    random.seed(s)
    xs = list(range(n))
    random.seed(30)
    random.shuffle(xs)
    return xs[:start_valid], xs[start_valid:]

def load_data(preprocessors=[], keep_df=False):
    random.seed(30)
    X = pd.read_csv('data/train.csv')
    for preprocessor in preprocessors:
        X = preprocessor(X)
    train_ids, valid_ids = shuffle_dataset(len(X))
    if keep_df:
        test_df = pd.read_csv('data/test.csv')
        return X, test_df
    Y = X[['label']]
    del X['label']
    train_X, train_Y = X.iloc[train_ids], Y.iloc[train_ids]
    valid_X, valid_Y = X.iloc[valid_ids], Y.iloc[valid_ids]
    return train_X, train_Y, valid_X, valid_Y
```

```python
from collections import Counter

def stats(df):
    xs = df[df.notna()]
    c = df.columns[0]
    print(f"min: {xs.min()[c]}, max: {xs.max()[c]}, mean {xs.mean()[c]}, median {xs.median()[c]}, std {xs.std()[c]}")
    
def stats_df(df):
    for c in df.columns:
        stats = Counter(df[c])
        print(f"{c}: {len(stats)} elements, tops 4: {stats.most_common(4)}")
```

```python
class DataBunch():
    def __init__(self, train_X, train_Y, valid_X, valid_Y):
        self.train_X, self.train_Y, self.valid_X, self.valid_Y = train_X, train_Y, valid_X, valid_Y
        self.assets = {}
        self.train_X = self.filling_na(self.train_X)
    def add_asset(self, k,v):
        self.assets[k] = v

    def filling_na(self, df):
        df = self._preprocess_FIELD_3(df)
        df = self._preprocess_FIELD_7(df)
        fillan_unknown_list = ['province', 'district', 'maCv', 'FIELD_1', 'FIELD_2', 'FIELD_4',
                               'FIELD_5', 'FIELD_6', 'FIELD_8', 'FIELD_9',
                               'FIELD_10', 'FIELD_11', 'FIELD_12', 'FIELD_13', 'FIELD_14', 'FIELD_15',
                              'FIELD_16', 'FIELD_17', 'FIELD_18', 'FIELD_19', 'FIELD_20', 'FIELD_21',
                              'FIELD_23', 'FIELD_24', 'FIELD_25', 'FIELD_26', 'FIELD_27', 'FIELD_28',
                              'FIELD_29', 'FIELD_30', 'FIELD_31', 'FIELD_32', 'FIELD_33', 'FIELD_34',
                              'FIELD_35', 'FIELD_36', 'FIELD_37', 'FIELD_38', 'FIELD_39', 'FIELD_40', 'FIELD_41',
                              'FIELD_42', 'FIELD_43', 'FIELD_44', 'FIELD_45', 'FIELD_46', 'FIELD_47',
                               'FIELD_48', 'FIELD_49', 'FIELD_50']
        fillan_median_list = ['age_source1', 'age_source2',
                              'FIELD_22', 'FIELD_51', 'FIELD_52', 'FIELD_53', 'FIELD_54', 'FIELD_55', 'FIELD_56', 'FIELD_57']
        df = self._fillna_unknown(df, fillan_unknown_list)
        df = self._fillna_median(df, fillan_median_list)
        return df
    
    def _fillna_median(self, df, columns):
        for c in columns:
            if c not in self.assets:
                median = df[[c]].median()[c]
                self.add_asset(c, median)
            df[[c]] = df[[c]].fillna(self.assets[c])
            stats(df[[c]])
        return df
    
    def _fillna_unknown(self, df, columns):
        for c in columns:
            if c not in self.assets:
                self.add_asset(c, 'unknown')
            df[[c]] = df[[c]].fillna(self.assets[c])
            stats_df(df[[c]])
        return df
    
    def _preprocess_FIELD_3(self, df, mode='-1'):
        c = 'FIELD_3'
        if c not in self.assets:
            if mode == '-1':
                self.add_asset(c, -1)

        df[[c]] = df[[c]].fillna(self.assets[c])
        stats(df[[c]])
        return df
    
    def _preprocess_FIELD_7(self, df):
        c = 'FIELD_7'
        if c not in self.assets:
            self.add_asset(c, [])
        df[[c]] = df[[c]].fillna(0)
        df[c] = df[c].apply(lambda x: x if x else [])
        return df
```
